```python
def f(x,b1,b2,b3):
    return b1*x+b2*x**2+b3*x**3
```

```python
popt, pcov=optimize.curve_fit(f,data["frequency [Hz]"].values,data['x [M_PNS/R_PNS**2]'].values,(10**(-6),-8*10**(-10),10**(-12)))
```

```python
fig, ax1 = plt.subplots(1,1, figsize=(6,6))
ax1.plot(data["frequency [Hz]"].values,data['x [M_PNS/R_PNS**2]'].values ,".r")
xx=np.linspace(0,1800,100)
ax1.plot(xx,f(xx,popt[0],popt[1],popt[2]), label="fit")
plt.plot(xx,f(xx,popt[0]+3*np.sqrt(np.abs(pcov[0,0])),popt[1]+3*np.sqrt(np.abs(pcov[1,1])),popt[2]+3*np.sqrt(np.abs(pcov[2,2]))),'--',label=(r"fit+3 $ \sigma$"))
plt.plot(xx,f(xx,popt[0]-3*np.sqrt(np.abs(pcov[0,0])),popt[1]-3*np.sqrt(np.abs(pcov[1,1])),popt[2]-3*np.sqrt(np.abs(pcov[2,2]))),'--',label=(r"fit-3 $\sigma$"))
plt.legend()
plt.xlabel('frÃ©quences(Hz)')
plt.ylabel('ratio r')
plt.title('r(f)')

```
